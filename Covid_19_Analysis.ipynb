```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cdata = pd.read_csv('/content/covid19_Confirmed_dataset.csv')
cdata.head(5)
```

```python
cdata.shape
```

```python
#deleting the useless columns
df = cdata.drop(['Lat','Long'],axis=1, inplace=True)
```

```python
cdata.head()
```

```python
#aggeregate rows by the country
c_data_agg = cdata.groupby('Country/Region').sum()
```

```python
c_data_agg.head(20)
```

```python
c_data_agg.shape
c_data_agg.drop(['Province/State'],axis=1,inplace=True)
```

```python
#visuallize data related to country
c_data_agg.loc['Afghanistan'].plot()
c_data_agg.loc['India'].plot()
c_data_agg.loc['Spain'].plot()
plt.legend()
```

```python
#calculate a good measure
c_data_agg.loc['India'][:30].plot()
```

```python
c_data_agg.loc['India'].diff().plot()
```

```python
#maximum infection rate
c_data_agg.loc['India'].diff().plot()
c_data_agg.loc['China'].diff().plot()
c_data_agg.loc['Spain'].diff().plot()
plt.legend()
```

```python
countries = list(c_data_agg.index )
max_infection_rate  = []
for c in countries:
  max_infection_rate.append(c_data_agg.loc[c].diff().max())
```

```python
c_data_agg['max_infection_rate'] = max_infection_rate
```

```python
c_data_agg.head()
```

```python
corona_data = pd.DataFrame(c_data_agg['max_infection_rate'])
```

```python
corona_data.head()
```

```python
#importing the happiness dataset
happiness_report = pd.read_csv('/content/worldwide_happiness_report.csv')
```

```python
happiness_report
```

```python
#drop the useless columns
useless_cols = ['Overall rank','Score','Generosity','Perceptions of corruption']
```

```python
happiness_report.drop(useless_cols,axis=1,inplace=True)
happiness_report.head()
```

```python
happiness_report.set_index('Country or region',inplace=True)
```

```python
happiness_report.head()
```

```python
#join the two dataset
corona_data.shape
```

```python
happiness_report.shape
```

```python
data = corona_data.join(happiness_report,how='inner')
data.head()
```
